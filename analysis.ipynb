```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
with open('output/expert_load.pkl', 'rb') as f:
    expert_load = pickle.load(f)
```

```python
# Total load
df = expert_load['total']

fig, ax = plt.subplots(figsize=(14, 8))
df.plot(kind='bar', stacked=True, ax=ax, width=1.0, edgecolor='none')

# Customize the plot
ax.set_xlabel('Iteration')
ax.set_ylabel('Expert Activations')
ax.set_title('Graph 0')
plt.legend(title='Expert', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xticks(range(0, len(df), 20))  # Show only every 20th tick
ax.set_xticklabels(df.index[::20], rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
